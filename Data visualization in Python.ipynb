# Import packages

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import networkx as nx
from wordcloud import WordCloud, STOPWORDS
import plotly
import plotly.plotly as py
import squarify
%matplotlib inline
```

# Word cloud using Standord University 2018 commencement speech

```python
f=open("Stanford 2018 commencement speech.txt","r")
file_lines_list = f.readlines()
f.close()

review_text = " ".join(file_lines_list)

wordcloud1 = WordCloud(
                          stopwords=STOPWORDS,
                          background_color='white',
                          min_font_size =6,
                          max_words=100,
                          width=2000,
                          height=2000
                      ).generate(review_text)

wordcloud_dict = wordcloud1.words_
wordcloud_dict.pop('one', None)
wordcloud_dict.pop('ve', None)
wordcloud_dict.pop('people', None)
wordcloud_dict.pop('re', None)
wordcloud_dict.pop('ll', None)
wordcloud_dict.pop('us', None)
wordcloud_dict.pop('know', None)
wordcloud_dict.pop('speech', None)
wordcloud_dict.pop('way', None)
wordcloud_dict.pop('something', None)
wordcloud_dict.pop('make', None)
wordcloud_dict.pop('best', None)
wordcloud_dict.pop('always', None)
wordcloud_dict.pop('thing', None)
wordcloud_dict.pop('now', None)
wordcloud_dict.pop('don', None)

wordcloud2 = WordCloud(
                          stopwords=STOPWORDS,
                          background_color='white',
                          min_font_size =6,
                          max_words=100,
                          width=5000,
                          height=5000
                         ).generate_from_frequencies(wordcloud_dict)
                       
plt.figure(figsize=(8,8))
plt.imshow(wordcloud2, interpolation="bilinear")
plt.axis('off')
plt.show()
```

# Social graphs plotted using the cosponsoring networks of female members of the House of Representative in Congress 105

**Note:** 

1. The four social graphs are generated using the same data but use different layouts. The upper left one is the Kamada-Kawai layout. The upper right one is a circular layout. The lower left one is the NEATO layout. The lower right one is the TWOPI layout.

2. The red nodes in the social graph represent female representatives from the Republican party, and the blue nodes in social graph represent female representatives from the Democratic Party. 

3. The red edges represent those that originated from Republican female representatives, and the blue edges represent those that are originated from Democratic female representatives.

```python
df = pd.read_stata('congress 105 women cosponsoring network.dta')

nrows = len(df.index)
sp_majority_dict = dict(zip(df.spID,df.sp_majority))
co_majority_dict = dict(zip(df.coID,df.co_majority))
sp_majority_dict.update(co_majority_dict)
politician_majority_dict = sp_majority_dict

G = nx.DiGraph()
for i in range(nrows):
    if politician_majority_dict[df.coID[i]] == 1:
        G.add_edges_from([(df.coID[i], df.spID[i], {'color': 'red'})])
    elif politician_majority_dict[df.coID[i]] == 0:
        G.add_edges_from([(df.coID[i], df.spID[i], {'color': 'blue'})])

edges = G.edges()
edge_color_list = [G[u][v]['color'] for u,v in edges]        
node_color_list = [] 
for n in G:
    if politician_majority_dict[n] == 1:
        node_color_list.append('red')
    else:
        node_color_list.append('blue')


plt.figure(figsize=(15,15))
plt.subplot(2, 2, 1)
pos = nx.kamada_kawai_layout(G)
nx.draw(G,pos=pos,with_labels=False,node_size=50,node_color=node_color_list,edge_color=edge_color_list,width=0.3)

plt.subplot(2, 2, 2)
pos = nx.nx_agraph.graphviz_layout(G, prog='circo')
nx.draw(G,pos=pos,with_labels=False,node_size=50,node_color=node_color_list,edge_color=edge_color_list,width=0.3)

plt.subplot(2, 2, 3)
pos = nx.nx_agraph.graphviz_layout(G, prog='neato')
nx.draw(G,pos=pos,with_labels=False,node_size=50,node_color=node_color_list,edge_color=edge_color_list,width=0.3)

plt.subplot(2, 2, 4)
pos = nx.nx_agraph.graphviz_layout(G, prog='twopi')
nx.draw(G,pos=pos,with_labels=False,node_size=50,node_color=node_color_list,edge_color=edge_color_list,width=0.3)
```

# Map showing GDP per capita in 2017

```python
df = pd.read_csv('world bank GDP per capita.csv')
#Please apply your own username and api_key at https://plot.ly/settings/api, 
#then replace the XXX with your own username and api_key
plotly.tools.set_credentials_file(username='XXX', api_key='XXX')

data = [ dict(
        type = 'choropleth',
        locations = df['Country Code'],
        z = df['2017 GDP per capita'],
        text = df['Country Name'],
        colorscale = "Reds",
        autocolorscale = False,
        reversescale = False,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            tickprefix = 'US$') )]

layout = dict(
    title = '<b>Country GDP per capita in 2017</b><br>Data source: World Bank',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='d3-world-map' )
```

# Tree map generated using Bansal & Roth (2000) Web of Science citation data

```python
#the two lines below are used to change the default text color and text font size in matplotlib
matplotlib.rcParams['text.color'] = 'white'
matplotlib.rcParams['font.size'] = 15

df = pd.read_csv('Bansal and Roth (2000) citation by journal.csv')
df["journal_citation"] = df['journal'] + "\n" + df['citation'].astype("str")

colors = ["red","darkgreen","navy","darkslategrey","darkorange","mediumvioletred",\
          "teal","olive","saddlebrown","limegreen","purple","dimgrey"]

plt.figure(figsize=(20,10))
squarify.plot(sizes=df['citation'], label=df['journal_citation'], color=colors, alpha=1 )
plt.axis('off')
plt.show()
```

# Bubble plot (only scatter plot, no variation for dot size) generated using Moro et al (2016) open data
**Note:** data source is https://archive.ics.uci.edu/ml/datasets/Facebook+metrics

```python
df = pd.read_csv('dataset_Facebook.csv')

plt.figure(figsize=(10,10))
plt.xlim(0,4000)
plt.ylim(0,70)
plt.scatter(df["Lifetime Engaged Users"], df["comment"], \
            color="blue", alpha=0.8)
plt.xlabel('Number of people who clicked the post',fontsize=15)
plt.ylabel('Number of comments on the post',fontsize=15)
plt.show()
```

# Bubble plot (hollow bubble) generated using Moro et al (2016) open data
**Note:** data source is https://archive.ics.uci.edu/ml/datasets/Facebook+metrics

```python
df = pd.read_csv('dataset_Facebook.csv')

plt.figure(figsize=(10,10))
plt.xlim(0,4000)
plt.ylim(0,70)

#the argument s represents the size of the bubble
plt.scatter(df["Lifetime Engaged Users"], df["comment"], s=df["like"], \
            color="white", edgecolors="blue", alpha=0.8)
plt.xlabel('Number of people who clicked the post',fontsize=15)
plt.ylabel('Number of comments on the post',fontsize=15)
plt.show()
```
